## 注释
以#开头、三个单引（'''）或双引（"""）号、一个也可以

```python
def rm29(x):
    '去除日数据闰年2.29'
    is_feb_29 = (x['time'].dt.month == 2) & (x['time'].dt.day == 29)
    x = x.sel(time=~is_feb_29)
    return x
```
